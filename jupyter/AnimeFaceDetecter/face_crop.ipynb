```python
import cv2
import os
import numpy as np
from time import sleep
```

```python
anime_cascade = cv2.CascadeClassifier('animeface/lbpcascade_animeface.xml')
```

```python
files = os.listdir(r'.\Datas\crol')
```

```python
def crop(arr,x,y,r):
    
    res = []
    for i in arr[y:y+r]:
        res.append(i[x:x+r])
    return np.array(res)


def f(_x):
    x, y, w, h = _x
    return w
```

```python
c = 1790
for file in files:
    img = cv2.imread(r'Datas\\crol\\' + file)
    gray = cv2.imread(r'Datas\\crol\\' + file,0)
    if gray is None:
        break
    if img is None:
        break
    cv2.imshow('img', gray)
    faces = anime_cascade.detectMultiScale(gray)
    if not type(faces) == tuple:
        faces = faces.tolist()
        faces.sort(key=f)
    for (x, y, w, h) in faces[:2]:
        if x and y and w and w > 60:
            c+=1
            print(file + str([x,y]))
            re_img = crop(img, x, y, w)
            cv2.imwrite(r"Datas\\result\\face\\{}.jpg".format(c), re_img)
            cv2.destroyAllWindows()
            re_img = crop(gray, x, y, w)
            cv2.imwrite(r"Datas\\result\\face_gray\\{}.jpg".format(c), re_img)
            cv2.destroyAllWindows()
            
sleep(5)
cv2.destroyAllWindows()
```
