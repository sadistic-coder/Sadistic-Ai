```python
import cv2
import os
import numpy as np
```

```python
anime_cascade = cv2.CascadeClassifier('animeface/lbpcascade_animeface.xml')
```

```python
files = os.listdir(r'C:\Users\haho6\Desktop\Datas\All')
```

```python
def crop(arr,x,y,r):
    
    res = []
    for i in arr[y:y+r]:
        res.append(i[x:x+r])
    return np.array(res)
```

```python
c = 356
for file in files:
    c+=1
    img = cv2.imread(r'Datas\\All\\' + file)
    gray = cv2.imread(r'Datas\\All\\' + file,0)
    if gray is None:
        break
    if img is None:
        break
    cv2.imshow('img', gray)
    faces = anime_cascade.detectMultiScale(gray)
    for (x, y, w, h) in faces:
        if x and y and w and w > 50:
            re_img = crop(img, x, y, w)

    cv2.imwrite(r"Datas\\result\\face\\{}.jpg".format(c), re_img)
    cv2.destroyAllWindows()
```
